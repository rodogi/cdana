# Outline

To prove: The cumulative intersection of nodes in main cores across different layers is not found at random. We propose two null-models:
1. One containing the number of nodes in main cores and select that number at random.
2. A second one containing the number of nodes found in a network with similar degree distribution.

# A generative model of the real distribution

This could be done by selecting a percentage of the number of nodes in the main cores.

# Compute cumulative intersection and jaccard similarity

Make a tidy dataframe with the cumulative intersection of the set of nodes in the networks.

```python
DATA = '/Users/rdora/breast_cancer/main_cores/'
FIG = '/Users/rdora/Dropbox/inmegen/breast_cancer/article/figures/'
```

```python
tissues = ['basal', 'healthy', 'her2', 'luma', 'lumb']
names = ['Basal', 'Healthy', 'Her2', 'Lum A', 'Lum B']
```

```python
main_cores = glob(join(DATA, 'healthy', '*'))
```

```python
dfs = []
for x, tissue in enumerate(tissues):
    main_cores = glob(join(DATA, tissue, '*'))
    main_cores = sorted(main_cores, key=lambda x: int(x.split('.')[0].split('/')[-1]), reverse=True)
    s1 = set(np.genfromtxt(main_cores[0], dtype='unicode'))
    s3 = set(np.genfromtxt(main_cores[0], dtype='unicode'))
    layer = 100
    cum_ints_ratio = [1.]
    sub_ints = [1.]
    cum_ints = [len(s1)]
    layers = [layer]
    for i, main_core in enumerate(main_cores[1:]):
        s2 = set(np.genfromtxt(main_core, dtype='unicode'))
        cum_ints.append(len(s1 & s2))
        if s1:
            cum_ints_ratio.append(len(s1 & s2) / len(s1))
        else:
            cum_ints_ratio.append(0)
        sub_ints.append(len(s2 & s3) / len(s3))
        s1 = s2 & s1
        s3 = s2
        layer -= 1
        layers.append(layer)
    df = pd.DataFrame({'Layer': layers,
                      'Cum_Int': cum_ints,
                       'Cum_Int_Ratio': cum_ints_ratio,
                      'Sub_Int': sub_ints})
    df['Tissue'] = names[x]
    dfs.append(df)
```

```python
df_tidy = pd.concat(dfs, axis=0)
```

```python
colors = ['#6aabd1', '#b63b2e', '#363441', '#ab9b93', '#abd16a']
pal = sns.color_palette(colors)

sns.palplot(pal)
```

```python
hue_order = ["Healthy",
            "Lum A",
            "Lum B",
            "Her2",
            "Basal"]
```

```python
g = sns.FacetGrid(df_tidy, col='Tissue', hue='Tissue', hue_order=hue_order,
                 palette=pal, col_order=hue_order)
g.map(sns.scatterplot, 'Layer', 'Sub_Int')
for ax in g.axes[0]:
    ax.invert_xaxis()
g.savefig(join(FIG, 'Subsequent_intersection.png'), dpi=300)
```

```python
g = sns.FacetGrid(df_tidy, col='Tissue', hue='Tissue', hue_order=hue_order,
                 palette=pal, col_order=hue_order)
g.map(sns.scatterplot, 'Layer', 'Cum_Int_Ratio')
for ax in g.axes[0]:
    ax.invert_xaxis()
g.savefig(join(FIG, 'Cumulative_intersection_ratio.png'), dpi=300)
```

```python
g = sns.FacetGrid(df_tidy, col='Tissue', hue='Tissue', hue_order=hue_order,
                 palette=pal, col_order=hue_order)
g.map(sns.scatterplot, 'Layer', 'Cum_Int')
for ax in g.axes[0]:
    ax.invert_xaxis()
g.savefig(join(FIG, 'Cumulative_intersection.png'), dpi=300)
```
