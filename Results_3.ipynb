## Data replacing

We are going to change the names of the parameters to some previously defined in other figures.

```python
toreplace = {'dmean': 'Mean Deg',
            'dstd': 'Std Deg',
            'ass': 'D.A',
            'ass_attr': 'A.A',
            'size': 'Size'}
sbm = sbm.replace(toreplace)
newcols = {'type': 'Tissue',
          'cutoff': 'Layer',
          'variable': 'Param',
          'value': 'Value'}
sbm = sbm.rename(columns=newcols)
layers = {'r-median': 'Median',
          'r-low': 'Low',
         'r-99-100': 'High'}
sbm = sbm.replace(layers)
newtissues = {'healthy': 'Healthy',
             'luma': 'Lum A',
             'lumb': 'Lum B',
             'her2': 'Her2',
             'basal': 'Basal'}
sbm = sbm.replace(newtissues)
```

# Visualization

We need a plot to show, for each parameter and layer, the distribution of its values in the SBMs together with the real value.

```python
colors = ['#6aabd1', '#b63b2e', '#363441', '#ab9b93', '#abd16a']
pal = sns.color_palette(colors)
sns.palplot(pal)
```

```python
orderparam = ['Mean Deg', 'Std Deg', 'D.A', 'A.A', 'Size']
ordertissue = ['Healthy', 'Lum A', 'Lum B', 'Her2', 'Basal']
xorder = ['Low', 'Median', 'High']
```

```python
g = sns.catplot(x="Layer", y="Value", hue="Tissue", row="Param", col="Tissue",
                palette=pal, data=sbm, sharey='row', alpha=0.5, order=xorder,
               row_order=orderparam, hue_order=ordertissue, legend_out=True)
axs = g.axes
rows = g.row_names
cols = g.col_names
hues = g.hue_names
for i, r in enumerate(rows):
    for j, c in enumerate(cols):
        t = df.loc[(df.Layer.isin([-1, 0, 100])) & (df.Tissue==c) &
                  (df.Param==r)].sort_values(by='Layer')
        ax = axs[i][j].scatter(axs[i][j].get_xticks(),
                               t['Value'].values, c='#2e8b57', zorder=3)
        
# plt.legend(handles=[plt.scatter([], [], c='#2e8b57', label='Real value')],
#          loc=(1.1, 3.6))
g.savefig(join(FIG, 'facet_grid_sbm.png'), dpi=300)
```
